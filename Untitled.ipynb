```python
import pandas_datareader.data as pdr
import fix_yahoo_finance as fix
import yfinance as yf
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
```

```python
class Data_fetcher():
    def __init__(self):
        pass

    def download_Data(self,start_date, end_date):
        data = yf.download("AAPL", start= start_date, end= end_date)
        self.data = data
    def preprocessing(self):
        self.y = []
        self.x = []
        fiting_temp = np.array(self.data.iloc[0: len(self.data), 1])
        scaler = MinMaxScaler()
        fiting_temp = scaler.fit_transform(fiting_temp.reshape(-1, 1))
        #fiting_temp = fiting_temp.reshape(-1, 1)
        for i in range((fiting_temp.shape[0] // 10) * 10 - 11):
            x = fiting_temp[np.arange(i,i+10),:]
            y = fiting_temp[i+10+1]
            self.x.append(x)
            self.y.append(y)
        self.y = np.array(self.y)
        self.x = np.array(self.x)
        train_start = 0
        train_end = int(self.y.shape[0]*0.8)
        test_start = train_end + 1
        test_end = self.y.shape[0]
        self.train_x = self.x[np.arange(train_start, train_end), :]
        self.train_y = self.y[np.arange(train_start, train_end), :]
        self.test_x = self.x[np.arange(test_start, test_end), :]
        self.test_y = self.y[np.arange(test_start, test_end), :]
    def shuffle_train(self):
        shuffle = np.arange(self.train_x.shape[0])
        np.random.shuffle(shuffle)
        self.train_x = self.x[shuffle, :]
        self.train_y = self.y[shuffle, :]
        print(self.train_y)
    def get_ytrain(self):
        return self.train_y
    def get_xtrain(self):
        return self.train_x
    def get_ytest(self):
        return self.test_y
    def get_xtest(self):
        return self.test_x
            
    
df = Data_fetcher()
df.download_Data(start_date = "2000-01-01", end_date = "2018-10-01")
df.preprocessing()
df.shuffle_train()
    
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(100, activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(1, activation=tf.nn.relu))

model.compile(optimizer="adam", loss="mean_squared_error")

model.fit(df.get_xtrain(), df.get_ytrain(), epochs=30)
model.evaluate(df.get_xtest(), df.get_ytest())
ar = model.predict(df.get_xtest())
print(ar[100])
print(df.get_ytest()[100])
```
